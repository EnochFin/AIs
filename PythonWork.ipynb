```python
import matplotlib.pyplot as plt #Import the plot library

def mean(sampleset): #definition header for the mean() function
    total=0
    for element in sampleset:
        total=total+element
    return total/len(sampleset)

myset2=[-75.,-100.,-75.,-100.,-75.]
myset=[2.,10.,3.,6.,4.,6.,10.] #We create the dataset(sampleset)
mymean=mean(myset2) #call mean function
plt.plot(myset2) #plots the set
plt.plot([mymean] * 7)#plot a line of 7 points to show the mean
```

```python
import math #This library is needed for pow

def mean(sampleset): #definition header for the mean() function
    total=0
    for element in sampleset:
        total=total+element
    return total/len(sampleset)

def variance(sampleset):
    total=0
    setmean=mean(sampleset)
    for element in sampleset:
        total=total+(math.pow(element-setmean,2))
    return total/len(sampleset)

myset1=[2.,10.,3.,6.,4.,6.,10.]
myset2=[-75.,-100.,-75.,-100.,-75.]
print("Variance of first set:" + str(variance(myset1)))
print("Variance of second set:" + str(variance(myset2)))
```

Bernoulli distributions(events and probability)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
distro = np.random.binomial(100, .6, 10000)
plt.hist(distro, 100, normed=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
distro = np.random.binomial(100, .6, 10000)/0.01
plt.hist(distro, 100, normed=1)
plt.show()
```

Logistic Distribution

```python
import matplotlib.pyplot as plt
import numpy as np

mu=0.5
sigma=0.5
distro2 = np.random.logistic(mu, sigma, 10000)
distro = np.random.normal(mu, sigma, 10000)
plt.hist(distro, 50, normed=1)
plt.hist(distro2, 50, normed=1)
plt.show()
```

Sigmoid curve

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()
cumulative = np.random.logistic(mu, sigma, 10000)
plt.hist(cumulative, 50, normed=1, cumulative=1)
plt.show()
```
